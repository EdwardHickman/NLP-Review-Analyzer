```python
#Task 2
df = pd.read_csv("com.easyfitness.csv")
list1 = []
for i in range(len(df.index)):
    text1=df.at[i,"Review"]
    text1 = "".join([char for char in text1 if char not in string.punctuation])
    text1 = "".join([char for char in text1 if char in string.printable])
    text2 = ""
    for char in text1:
        if char in string.digits:
            char = char.replace(char,num2words(char))
            text2+=char
        else:
            text2+=char
    text1=text2
    text1 = " ".join(text1.split())
    text1=text1.lower()
    word_tokens = word_tokenize(text1)
    stop_words = set(stopwords.words('english'))
    text1 = [w for w in word_tokens if not w.lower() in stop_words]
    lemmatizer = WordNetLemmatizer()
    for j in range(len(df.index)):
        text1 = [lemmatizer.lemmatize(word) for word in text1]
    list1.append(text1)
    
```

```python
#Task 2 Bonus: I think that the choice of stop words matters because it can change the sentiment of the review. If certain words are removed from a negative review, it can seem like
#a positive one due to the fact that both negative and positive reviews will have a lot of words in common. For example, if two reviews are "there are many features" and 
#"there are not many features", removing stop words might make these reviews the same.

#Task 3
pd.set_option('display.max_rows', 250) #replace n with the number of rows you want to see completely
pd.options.display.max_colwidth = 100
df = pd.read_csv("com.bodybuilding.rise.csv")
df["Polarity"] = ''
df = df.loc[:, ['Package name','Review','Polarity']]
for i in range(250):
    b = TextBlob(" ".join(list6[i]))
    df["Polarity"][i] = b.sentiment
df.drop(df.index[250:], inplace=True)
display(df)
```

```python
analyzer = SentimentIntensityAnalyzer()
for j in range(250):
    sentences = " ".join(list6[j])
    df["Polarity"][j] = analyzer.polarity_scores(sentences)
display(df)
```
